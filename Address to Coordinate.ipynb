```python
import geopandas as gpd
from geopy.geocoders import Nominatim
import json
import csv
import pandas as pd
import numpy as np
```

```python
blockchain_data = pd.read_csv("./data/blockchain_data.csv")
tdg_data = pd.read_csv("./data/tdg_data.csv")
h2well_data = pd.read_csv("./data/h2well_data.csv")
campfire_data = pd.read_csv("./data/campfire_data.csv")
```

```python
geolocator = Nominatim(user_agent="gadm_matcher")

# Function to geocode an address
def geocode_address(address):
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

def iter_coords(func):
    def wrapper(*args, **kwargs):
        try:
            func()
        except Exception as e:
            print(e)
            return wrapper(func)
    return wrapper

def format_addr(addr):
    "Formats address for geocoding"
    return addr
```

```python
blockchain_addr = blockchain_data["person_address"].unique()
blockchain_coords = {}
blockchain_fails = []
```

```python
@iter_coords
def x():
    for addr in blockchain_addr[:]:
        if addr in blockchain_coords.keys():
            continue # Don't call API twice, they like to limit you
        addr = format_addr(addr)
        coord = geocode_address(addr)
        if coord[0] is None:
            print("FAILED for", addr)
        else:
            print(coord, addr)
            blockchain_coords[addr] = coord

x()
```
